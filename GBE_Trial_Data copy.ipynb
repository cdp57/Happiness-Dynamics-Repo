```python
import scipy.io 
import pandas as pd
mat = scipy.io.loadmat('Rutledge_GBE_risk_data_TOD.mat')

array = mat['subjData']

inner = array[0]

table = {
    'participant': [],
    'age': [],
    'gender': [],
    'attempt number': [],
    'time of day': [],
    'design version': [],
    'trial number': [],
    'screen side': [],
    'safe value': [],
    'winning value': [],
    'losing value': [],
    'choice made': [],
    'trial outcome': [],
    'choice latency': [],
    'happiness rating': [],
    'cursor starting point': [],
    'answer time taken': [],
    'spinner time taken': [],
    'spinner angle': [],
}

participant_number = 0

while participant_number < 47067:  

    times_completed = 0

    person = inner[participant_number]

    data = person['data']

    while len(data[0]) > times_completed:

        current_trial = 0

        while current_trial < 30:

            table['participant'].append(person['id'][0][0])
            table['age'].append(person['age'][0][0])
            table['gender'].append(person['isFemale'][0][0])
            table['attempt number'].append(times_completed + 1)
            table['time of day'].append(person['timeOfDay'][0][times_completed])
            table['design version'].append(person['designVersion'][0][times_completed])
            table['trial number'].append(data[0][times_completed][current_trial][0])
            table['screen side'].append(data[0][times_completed][current_trial][1])
            table['safe value'].append(data[0][times_completed][current_trial][2])
            table['winning value'].append(data[0][times_completed][current_trial][3])
            table['losing value'].append(data[0][times_completed][current_trial][4])
            table['choice made'].append(data[0][times_completed][current_trial][6])
            table['trial outcome'].append(data[0][times_completed][current_trial][7])
            table['choice latency'].append(data[0][times_completed][current_trial][8])
            table['happiness rating'].append(data[0][times_completed][current_trial][9])
            table['cursor starting point'].append(data[0][times_completed][current_trial][10])
            table['answer time taken'].append(data[0][times_completed][current_trial][11])
            table['spinner time taken'].append(data[0][times_completed][current_trial][13])
            table['spinner angle'].append(data[0][times_completed][current_trial][14])

            current_trial += 1

        times_completed += 1

    participant_number += 1

#data frame for trial results (provided values only, no calculations)
df = pd.DataFrame(table)
display(df)
```

```python
#updated data frame containing trials with design versions 1 and 2 only, results for certain value, expected value, and predicton error also added
df2 = df[df['design version'] < 2]
certain_val = []
expected_val = []
prediction_error = []
for index in df2.index:
    if df2['choice made'][index] == 0:
        certain_val.append(df2['safe value'][index])
        expected_val.append(0)
    else:
        certain_val.append(0)
        expected_val.append((df2['winning value'][index] + df2['losing value'][index]) / 2)

df2.insert(19, 'certain reward', certain_val)
df2.insert(20, 'expected value', expected_val)

for index in df2.index:
    if df2['choice made'][index] == 1:
        prediction_error.append(df2['trial outcome'][index] - df2['expected value'][index])
    else:
        prediction_error.append(0)

df2.insert(21, 'prediction error', prediction_error)

n = 30

#split DataFrame into chunks
list_df = [df2[i:i+n] for i in range(0, len(df2), n)]
```

```python
#add weighted summation values for CR, EV, PE
import math
gamma4 = 0.4
gamma6 = 0.6
gamma8 = 0.8
CR_gamma4 = []
CR_gamma6 = []
CR_gamma8 = []
EV_gamma4 = []
EV_gamma6 = []
EV_gamma8 = []
PE_gamma4 = []
PE_gamma6 = []
PE_gamma8 = []
H_gamma4 = []
H_gamma6 = [] 
H_gamma8 = []
for subframe in list_df:
    start = subframe.index[0]
    for i in subframe.index:
        sumCR4 = 0
        sumCR6 = 0
        sumCR8 = 0
        sumEV4 = 0
        sumEV6 = 0
        sumEV8 = 0
        sumPE4 = 0
        sumPE6 = 0
        sumPE8 = 0
        sumH4 = 0
        sumH6 = 0
        sumH8 = 0
        for j in range(start, i+1):
            sumCR4 += subframe['certain reward'][j] * gamma4**(i-j)
            sumCR6 += subframe['certain reward'][j] * gamma6**(i-j)
            sumCR8 += subframe['certain reward'][j] * gamma8**(i-j)
            sumEV4 += subframe['expected value'][j] * gamma4**(i-j)
            sumEV6 += subframe['expected value'][j] * gamma6**(i-j)
            sumEV8 += subframe['expected value'][j] * gamma8**(i-j)
            sumPE4 += subframe['prediction error'][j] * gamma4**(i-j)
            sumPE6 += subframe['prediction error'][j] * gamma6**(i-j)
            sumPE8 += subframe['prediction error'][j] * gamma8**(i-j) 
            if(math.isnan(subframe['happiness rating'][j]) == False):
                sumH4 += subframe['happiness rating'][j] * gamma4**(i-j)
                sumH6 += subframe['happiness rating'][j] * gamma6**(i-j)
                sumH8 += subframe['happiness rating'][j] * gamma8**(i-j)
        CR_gamma4.append(sumCR4)
        CR_gamma6.append(sumCR6)
        CR_gamma8.append(sumCR8)
        EV_gamma4.append(sumEV4)
        EV_gamma6.append(sumEV6)
        EV_gamma8.append(sumEV8)
        PE_gamma4.append(sumPE4)
        PE_gamma6.append(sumPE6)
        PE_gamma8.append(sumPE8)
        H_gamma4.append(sumH4)
        H_gamma6.append(sumH6)
        H_gamma8.append(sumH8)
df2.insert(22, 'CR gamma4', CR_gamma4)
df2.insert(23, 'CR gamma6', CR_gamma6)
df2.insert(24, 'CR gamma8', CR_gamma8)
df2.insert(25, 'EV gamma4', EV_gamma4)
df2.insert(26, 'EV gamma6', EV_gamma6)
df2.insert(27, 'EV gamma8', EV_gamma8)
df2.insert(28, 'PE gamma4', PE_gamma4)
df2.insert(29, 'PE gamma6', PE_gamma6)
df2.insert(30, 'PE gamma8', PE_gamma8)
df2.insert(31, 'H gamma4', H_gamma4)
df2.insert(32, 'H gamma6', H_gamma6)
df2.insert(33, 'H gamma8', H_gamma8)
```

```python
cos2 = []
cos4 = []
cos6 = []
cos8 = []
for index in df2.index:
    cos2.append(math.cos((df2['time of day'][index] * 24) / 2))
    cos4.append(math.cos((df2['time of day'][index] * 24) / 4))
    cos6.append(math.cos((df2['time of day'][index] * 24) / 6))
    cos8.append(math.cos((df2['time of day'][index] * 24) / 8))

df2.insert(34, 'cos2', cos2)
df2.insert(35, 'cos4', cos4)
df2.insert(36, 'cos6', cos6)
df2.insert(37, 'cos8', cos8)
```

Variables

participant: id of particpant from 1-47067

attempt number: the number of the participant's current attempt

trial number: 0-30 (each attempt consists of 30 trials)

time of day: 0-1, indicates fraction of the day elapsed

design version: 0-5, indicates version of the app

screen side: side of screen with risky option, 1 (left) or 2 (right)

safe value: value of safe option

winning value: value on "winning" side of spinner

losing value: value of "losing" side of the spinner

choice made: 1 (gamble), 0 (safe)

trial outcome: value the participant actually recieved 

choice latency: time taken to choose gamble or safe option

happiness rating: 0-100 if player asked, nan if player not asked

cursor starting point: 0-100 indicating the starting position of the cursor on the happiness scale, nan if not asked
***started at 50 for versions 1-2, started between 25-75 for versions 3-5

answer time taken: time taken to answer happiness question, nan if not asked

spinner time taken: amount of time taken for spinner to indicate outcome, versions 3-5 use negative values for counterclockwise spinning

spinner angle: angle the spinner stopped at, less than pi for right side, greater than pi for left

certain reward: the safe value if it is chosen, 0 if the gamble is chosen

expected value: expected outcome of the gamble if it is chosen, 0 if safe option is chosen

prediction error: outcome - expected value if gamble is chosen, 0 if safe option is chose

CR gamma 4/6/8: weighted summation of certain values using forgetting factor with gamma set to .4, .6, .8

EV gamma 4/6/8: weighted summation of expected values using forgetting factor with gamma set to .4, .6, .8

PE gamma 4/6/8: weighted summation of prediction errors using forgetting factor with gamma set to .4, .6, .8

H gamma 4/6/8: weighted summation of happiness ratings using forgetting factor with gamma set to .4, .6, .8

cos 2/4/6/8: cosine values of the number of hours elapsed in a day divided by a constant
