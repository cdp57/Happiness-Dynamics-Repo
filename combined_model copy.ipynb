```python
#weighted cubic calculations
data = pd.read_csv('GBE_final.csv')
data = data[data['attempt number'] < 2]

list_df = []

for i in range (1, 50000):
    df1 = data[data['participant'] == i]
    if len(df1) > 0:
        list_df.append(df1)


wCl4, wCl6, wCl8, wCc4, wCc6, wCc8 = [], [], [], [], [], []

for subframe in list_df:
    start = subframe.index[0]
    for i in subframe.index:
        sum1, sum2, sum3, sum4, sum5, sum6 = 0, 0, 0, 0, 0, 0
        
        for j in range(start, i+1):
            sum1 += subframe['Cl1'][j] * 0.4**(i-j)
            sum2 += subframe['Cl1'][j] * 0.6**(i-j)
            sum3 += subframe['Cl1'][j] * 0.8**(i-j)
            sum4 += subframe['Cc1'][j] * 0.4**(i-j)
            sum5 += subframe['Cc1'][j] * 0.6**(i-j)
            sum6 += subframe['Cc1'][j] * 0.8**(i-j)
        
            
        wCl4.append(sum1)
        wCl6.append(sum2)
        wCl8.append(sum3)
        wCc4.append(sum4)
        wCc6.append(sum5)
        wCc8.append(sum6)
   
        
data['wCl4'] = wCl4
data['wCl6'] = wCl6
data['wCl8'] = wCl8
data['wCc4'] = wCc4
data['wCc6'] = wCc6
data['wCcc8'] = wCc8


data.to_csv("GBE_final2.csv", index=False, header=True)
```
